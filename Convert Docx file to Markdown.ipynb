# **Loop Through All Files in the mass_convert directory**

```base="${file%.docx}"``` means: **take the filename stored in** ```$file``` and **remove the** ```.docx``` **suffix from the end.**

````bash
BASE_DIR="mass_convert"

echo "Scanning folders in $BASE_DIR"

cd "$BASE_DIR" || exit 1

for file in *.docx; do

    base="${file%.docx}"
    echo "Original: $file → Base: $base"

    pandoc -t gfm --extract-media . "$base.docx" -o "$base.md"
    
done
````
