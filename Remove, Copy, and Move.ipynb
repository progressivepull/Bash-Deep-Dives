# 📄 Copying files and folders (cp)
**cp** duplicates files, and **cp -r** copies directories recursively.

```bash
echo -e "# John Notes\nThe bear went to the store." > notes.txt
```

```bash
# Copy a file
cp notes.txt backup_notes.txt
```

```bash
ls
```

```bash
rm backup_notes.txt
```

```bash
rm notes.txt
```

```bash
ls
```
