```bash
# 📁 Bash snippet: put full folder paths into an array
```

```bash
# Get absolute path of this directory (optional, kept for context)
ROOT_DIR="$(pwd)"
```

```bash
echo "$ROOT_DIR"
```

```bash
# Array to hold folder names
folders=()
```

```bash
# Collect only directory names in current directory
for d in */ ; do
    # '*/' matches only directories; strip trailing slash
    folders+=("${d%/}")
done
```

```bash
# Print each element with its index
for (( i=0; i<${#folders[@]}; i++ )); do
    echo "Index $i: ${folders[$i]}"
done
```
