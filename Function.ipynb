# 📌 Clean, practical Bash function examples

## 1. Function with no arguments

```bash
say_hello() {
    echo "Hello from the function!"
}

say_hello
```

## 2. Function that takes arguments

```bash
print_folder() {
    echo "Folder name: $1"
}

print_folder "mnt"
print_folder "pinkt"
```

* **$1** = first argument

* **$2** = second argument

* **$@** = all arguments

```bash
print_folder() {
    echo "Folder name: $1"
    echo "Folder name: $2"
    echo "Folder name: $@"
}

print_folder "mnt"  "pinkt"
```

## 3. Function that checks if a directory exists (if/else inside)

```bash
check_dir() {
    if [[ -d "$1" ]]; then
        echo "Directory exists: $1"
    else
        echo "Directory does NOT exist: $1"
    fi
}

check_dir "media"
check_dir "not_here"
```

**This uses the directory test inside a function.**

# 4. Function that loops through an array

```bash
print_array() {
    for i in "${!folders[@]}"; do
        echo "Index $i: ${folders[$i]}"
    done
}

folders=("mnt" "d" "pinkt")
print_array
```

Uses **bash array indexing** and **bash array looping**.

## 5. Function that returns a value (via echo)
Bash functions don’t return values like other languages; **you “return” by printing**:

```bash
get_current_dir_name() {
    echo "$(basename "$(pwd)")"
}

dir_name="$(get_current_dir_name)"
echo "Current directory: $dir_name"
```

Uses **basename** to extract the last path component.
