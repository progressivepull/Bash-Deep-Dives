**Create three markdown files with a header and paragraph**

**echo -e** tells Bash to interpret backslash escape sequences inside the quoted text. Without **-e**, Bash prints the characters literally.

**What echo -e does**
echo -e enables special sequences such as:

* \n → newline

* \t → tab

* \\ → literal backslash

* \r → carriage return

* \b → backspace

**Note**         
**\b** represents a backspace escape sequence, meaning it deletes the previous character when interpreted by a command like **echo -e**.


```bash
echo -e "# File One\nThis is the first markdown file." > file1.md
echo -e "# File Two\nThis is the second markdown file." > file2.md
echo -e "# File Three\nThis is the third markdown file." > file3.md
```
