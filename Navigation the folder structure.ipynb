# EXAMPLE:  Get the full directory path, parent directory and last path

* **pwd** prints the full working directory path

* **dirname** returns the parent directory

* **basename** extracts the last path component

```bash
PATH_HOME="/home/user/projects/demo"
```

```bash
echo "$PATH_HOME"
```

```bash
CURRENT_DIR_HOME="$(basename $PATH_HOME)"
```

```bash
echo "$CURRENT_DIR_HOME"
```

```bash
PARENT_DIR_HOME="$(dirname $PATH_HOME)"
```

```bash
echo "$PARENT_DIR_HOME"
```
