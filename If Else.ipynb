# 📌 Simple if/else example

````bash
num=10

if [[ $num -gt 5 ]]; then
    echo "Number is greater than 5"
else
    echo "Number is 5 or less"
fi
````

* ```[[ $num -gt 5 ]]``` checks if the number is greater than 5.

* The first block runs when the condition is true.

* The else block runs otherwise.

# 📁 Directory check example

````bash
dir="media"

if [[ -d "$dir" ]]; then
    echo "Directory exists: $dir"
else
    echo "Directory does not exist: $dir"
fi
````

This uses the directory test to check if a folder exists.

# 🔤 String comparison example

````bash
name="pinkt"

if [[ "$name" == "pinkt" ]]; then
    echo "Match!"
els
    echo "No match."
fi
````

# 🔁 if / elif / else example

````bash
value=2

if [[ $value -eq 1 ]]; then
    echo "Value is 1"
elif [[ $value -eq 2 ]]; then
    echo "Value is 2"
else
    echo "Value is something else"
fi
````
