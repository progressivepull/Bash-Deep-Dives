# 📦 Creating a Bash array

```bash
folders=("mnt" "d" "pinkt")
```

This stores three string elements at indices 0, 1, and 2.

# 🔍 Accessing elements

```bash
echo "Index 0: ${folders[0]}"
echo "Index 1: ${folders[1]}"
echo "Index 2: ${folders[2]}"
```

# 🔁 Looping through values

```bash
for f in "${folders[@]}"; do
    echo "Folder: $f"
done
```

```bash
echo "${folders[@]}"
```

# 🔢 Looping through indices

```bash
for i in "${!folders[@]}"; do
    echo "Index $i: ${folders[$i]}"
done
```

# 🔧 Adding elements

```bash
folders+=("new_folder")
```

```bash
echo "${folders[@]}"
```
